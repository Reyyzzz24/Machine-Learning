#Web Scraping

web scraping adalah proses secara otomatis untuk mengunduh/mengambil informasi dari halaman website, teknik ini menggunakan program untuk membaca konten website kemudian mengambil data yang diinginkan, seperti gambar, teks, tabel didalam format HTML.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_countries_by_real_GDP_growth_rate'
page = requests.get(url)

soup = BeautifulSoup(page.text,'html')
```

```python
soup
```

```python
table = soup.find_all('table', class_ = 'wikitable sortable static-row-numbers static-row-header-text')[2] # mengambil table ke 3

table
```

```python
table.find_all('th')
```

```python
judul_kolom = table.find_all('th') #ambil bagian judul kolom saja
list_judul_kolom = [judul.text for judul in judul_kolom]
print(list_judul_kolom)
```

```python
list_judul_kolom = [judul.text.strip() for judul in judul_kolom]
print(list_judul_kolom)
```

```python
import pandas as pd
df = pd.DataFrame(columns= list_judul_kolom)
df
```

```python
data_kolom = table.find_all('tr')
```

```python
for baris in data_kolom:
  data_baris = baris.find_all('td')
  data_baris_individu = [data.text.strip() for data in data_baris]
  print(data_baris_individu)
```

```python
for baris in data_kolom[1:]: #baris dimulai dari 1 dan seterusnya
  data_baris = baris.find_all('td')
  data_baris_individu = [data.text.strip() for data in data_baris]
  print(data_baris_individu)
  #pindah data ke data frame
  panjang = len(df)
  df.loc[panjang] = data_baris_individu
```

```python
df
```

```python
import numpy as np

df.replace(['n/a','na'], np.nan, inplace=True) #ubah menjadi NaN (Not a Number)
```

```python
df['2013'][0]
```

```python
type(df['2013'][0])
```

```python
for kolom in df.columns[1:]: #karena negara tidak dijadikan float
  df[kolom] = pd.to_numeric(df[kolom], errors='coerce').astype('float64')

df
```

```python
for kolom in df.columns[1:]:
  df[kolom].fillna(0, inplace=True)
  #df[kolom].fillna(df[kolom].mean(), inplace=True) dicari rata-rata
df
```

```python
#Metode Regresi

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

x = np.array(df['2022']).reshape(-1,1) #data input
y = np.array(df['2023']) #output

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)

model = LinearRegression()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print(f'hasil MSE = {mse}')
```
