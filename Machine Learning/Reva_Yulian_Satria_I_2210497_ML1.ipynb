```python
import pandas as pd
data = {"nama": ['andi','mulyono','budi'],
        "umur": [21,30,35],
        "kota": ['jakarta','solo', 'bogor']}
data_2 = pd.DataFrame(data)
print(data_2)
```

```python
display(data_2)
```

```python
display(data_2[data_2.nama != 'andi'])
```

```python
display(data_2[(data_2.nama == 'mulyono') | (data_2.nama == 'budi')])
```
